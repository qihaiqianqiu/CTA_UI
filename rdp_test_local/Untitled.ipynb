```python
import os
import win32clipboard
import subprocess
```

```python
def linux_to_windows(path):
    return path.replace("/", "\\")
```

```python
local_file_path = os.path.join(r"D:\local_repo\CTA_UI", "rdp_test_local", "params.csv")
remote_file_path = os.path.join(r"C:\Users\Administrator\Desktop", "rdp")
rdp_clip_exe = r"C:/Windows/System32/rdpclip.exe"
```

```python
import subprocess
remote_file_path = linux_to_windows(remote_file_path)
```

```python
with open(local_file_path, 'r', encoding='utf-8') as f:
    file_data = f.read()
```

```python
file_data
```

```python
win32clipboard.OpenClipboard()
win32clipboard.EmptyClipboard()
win32clipboard.SetClipboardData(win32clipboard.CF_UNICODETEXT, file_data)
```

```python
win32clipboard.GetClipboardData(win32clipboard.CF_UNICODETEXT)
```

```python
win32clipboard.CloseClipboard()
```

```python
subprocess.run(f"cmd /c echo.|set /p=\"{remote_file_path}\"|clip", shell=True)
```
